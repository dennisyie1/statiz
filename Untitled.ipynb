```python
def team_json(team_code, date, show=False): # show = True일때만 데이터 보여줌.

    METHOD = "GET"
    PATH   = "prediction/playerRoster"
    QUERY  = {
        "date": str(date),   # ← 날짜 인자로 받기
        "code": "1",
        "t_code": str(team_code)
    }

    def normalize_query(params: dict) -> str:
        safe = "-_.!~*'()"
        return "&".join(
            f"{quote(str(k), safe=safe)}={quote(str(params[k]), safe=safe)}"
            for k in sorted(params.keys())
        )

    def make_signature(secret: str, payload: str) -> str:
        return hmac.new(
            SECRET.encode("utf-8"),
            payload.encode("utf-8"),
            hashlib.sha256
        ).hexdigest()

    normalized = normalize_query(QUERY)
    timestamp = str(int(time.time()))

    payload = f"{METHOD}|{PATH}|{normalized}|{timestamp}"
    signature = make_signature(SECRET, payload)

    url = f"https://api.statiz.co.kr/baseballApi/{PATH}?{normalized}"

    headers = {
        "X-API-KEY": API_KEY,
        "X-TIMESTAMP": timestamp,
        "X-SIGNATURE": signature,
    }

    req = urllib.request.Request(url, method=METHOD, headers=headers)

    with urllib.request.urlopen(req, timeout=30) as resp:
        body = resp.read().decode("utf-8")

    json_data = json.loads(body)

    # 옵션 출력
    if show:
        display(json_data)
        print(json_data.keys())

    return json_data
```

## JSON 파일로 저장 완료

```python
def preprocess_hitter_json(player_id, file_path=None):
    """
    타자 1명의 playerSeason JSON 파일을 불러와
    basic + deepen을 병합하고 dtype을 정리한 DataFrame을 반환한다.
    
    Parameters
    ----------
    player_id : int or str
        선수 ID
    file_path : str, optional
        JSON 파일 경로. 지정하지 않으면 "{player_id}.json" 사용
    
    Returns
    -------
    pd.DataFrame
        전처리 완료된 타자 시즌별 데이터프레임
    """
    
    if file_path is None:
        file_path = f"{player_id}.json"
    
    # JSON 불러오기
    with open(file_path, "r", encoding="utf-8") as f:
        json_data = json.load(f)
    
    # DataFrame 변환
    basic_df = pd.DataFrame(json_data["basic"]["list"])
    deepen_df = pd.DataFrame(json_data["deepen"]["list"])
    
    # 타자 JSON에서는 deepen에 year가 없는 경우가 있으므로 basic_df의 year를 붙임
    deepen_df["year"] = basic_df["year"].values
    
    # basic + deepen 병합
    merged_df = basic_df.merge(
        deepen_df[["year", "BBK", "BABIP", "wOBA"]],
        on="year",
        how="left"
    )
    
    # 컬럼 타입 정리
    str_cols = ["p_name", "p_nameEN", "t_code"]
    
    int_cols = [
        "p_no", "leagueType", "year", "p_position",
        "G", "PA", "ePA", "AB", "R", "H", "2B", "3B", "HR", "TB",
        "RBI", "SB", "CS", "BB", "HP", "IB", "SO", "GDP", "SH", "SF"
    ]
    
    float_cols = [
        "WAROff", "AVG", "OBP", "SLG", "OPS", "wRCplus", "WAR",
        "BBK", "BABIP", "wOBA"
    ]
    
    for c in str_cols:
        if c in merged_df.columns:
            merged_df[c] = merged_df[c].astype("object")
    
    for c in int_cols:
        if c in merged_df.columns:
            merged_df[c] = pd.to_numeric(merged_df[c], errors="coerce").astype("Int64")
    
    for c in float_cols:
        if c in merged_df.columns:
            merged_df[c] = pd.to_numeric(merged_df[c], errors="coerce")
    
    return merged_df
```

```python
def preprocess_pitcher_json(player_id, file_path=None):
    """
    투수 1명의 playerSeason JSON 파일을 불러와
    basic + deepen을 병합하고 dtype을 정리한 DataFrame을 반환한다.
    
    Parameters
    ----------
    player_id : int or str
        선수 ID
    file_path : str, optional
        JSON 파일 경로. 지정하지 않으면 "{player_id}.json" 사용
    
    Returns
    -------
    pd.DataFrame
        전처리 완료된 투수 시즌별 데이터프레임
    """
    
    if file_path is None:
        file_path = f"{player_id}.json"
    
    with open(file_path, "r", encoding="utf-8") as f:
        json_data = json.load(f)
    
    basic_df = pd.DataFrame(json_data["basic"]["list"])
    deepen_df = pd.DataFrame(json_data["deepen"]["list"])
    
    deepen_df["year"] = basic_df["year"].values
    
    merged_df = basic_df.merge(
        deepen_df[["year", "K9", "BB9", "KBB", "HR9", "OBP", "SLG", "OPS"]],
        on="year",
        how="left"
    )
    
    # ===== IP 변환 함수 =====
    def convert_ip(ip_value):
        if pd.isna(ip_value):
            return ip_value
        
        ip_str = str(ip_value)
        
        if "." not in ip_str:
            return float(ip_value)
        
        integer, decimal = ip_str.split(".")
        
        if decimal == "1":
            return float(integer) + 0.3
        elif decimal == "2":
            return float(integer) + 0.6
        else:
            return float(ip_value)
    
    # ===== dtype 처리 =====
    str_cols = ["p_name", "p_nameEN", "t_code"]
    
    int_cols = [
        "p_no", "leagueType", "year", "p_position",
        "G", "GS", "GR", "GF", "CG", "SHO", "W", "L", "S", "HD",
        "ER", "R", "TBF",
        "H", "2B", "3B", "HR",
        "BB", "HP", "IB", "SO",
        "ROE", "BK", "WP",
        "QS"
    ]
    
    float_cols = [
        "IP", "ERA", "FIP", "WHIP", "WAR",
        "K9", "BB9", "KBB", "HR9",
        "OBP", "SLG", "OPS"
    ]
    
    for c in str_cols:
        if c in merged_df.columns:
            merged_df[c] = merged_df[c].astype("object")
    
    for c in int_cols:
        if c in merged_df.columns:
            merged_df[c] = pd.to_numeric(merged_df[c], errors="coerce").astype("Int64")
    
    for c in float_cols:
        if c in merged_df.columns:
            merged_df[c] = pd.to_numeric(merged_df[c], errors="coerce")
    
    # ===== IP 변환 적용 =====
    if "IP" in merged_df.columns:
        merged_df["IP_converted"] = merged_df["IP"].apply(convert_ip)
    
    return merged_df
```

```python
def preprocess_player_json(player_id, file_path=None):
    """
    선수 JSON 파일을 읽어 p_position 기준으로
    투수면 preprocess_pitcher_json(),
    타자면 preprocess_hitter_json()
    을 적용해 반환한다.
    
    Parameters
    ----------
    player_id : int or str
        선수 ID
    file_path : str, optional
        JSON 파일 경로. 지정하지 않으면 "{player_id}.json" 사용
    
    Returns
    -------
    pd.DataFrame
        투수 또는 타자 전처리 함수의 반환값
    """
    
    if file_path is None:
        file_path = f"{player_id}.json"
    
    # 원본 JSON 읽기
    with open(file_path, "r", encoding="utf-8") as f:
        json_data = json.load(f)
    
    # basic의 첫 row에서 포지션 확인
    basic_list = json_data.get("basic", {}).get("list", [])
    
    if not basic_list:
        raise ValueError(f"basic.list가 비어 있습니다: {file_path}")
    
    p_position = basic_list[0].get("p_position", None)
    
    if p_position is None:
        raise ValueError(f"p_position 정보를 찾을 수 없습니다: {file_path}")
    
    # 투수 / 타자 분기
    if int(p_position) == 1:
        return preprocess_pitcher_json(player_id, file_path=file_path)
    else:
        return preprocess_hitter_json(player_id, file_path=file_path)
```
